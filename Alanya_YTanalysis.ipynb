## 3. For each channel, determine the total number of views

By inspecting the dataset, we observe that the `views` column is cumulative. A video can remain in the trending section for multiple consecutive days, creating a new row for each day it trends within a specific country. If we were to simply use `.groupby('channel_title')['views'].sum()`, we would sum the views of the same video multiple times, leading to heavily inflated and incorrect totals.


Look at Great Britain (GB): the video trended from November 14th all the way to November 27th, and the views grew from 33 million to over 138 million.

```python
df[(df['video_id']=='2Vv-BfVoq4g')&(df['country'] == 'GB')][['trending_date','title','views','country']]
```

To obtain the correct number of total views for each channel, we implement the following strategy:

1. We convert `trending_date` into a proper datetime object 

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format = '%y.%d.%m')
```

2. We sort the DataFrame by `country` and `trending_date` in ascending order. 

```python
df_sorted = df.sort_values(by=['country', 'trending_date'])
df_sorted.head()
```

3. Since views are cumulative, the last occurrence of a video in a given country contains its final and maximum view count.

```python
df_unique_videos = df_sorted.drop_duplicates(subset = ['video_id', 'country'], keep= 'last')
df_unique_videos.head()
```

4. Finally, we group the deduplicated data by `channel_title` and sum the unique final views.

```python
channel_views = df_unique_videos.groupby('channel_title', as_index=False)['views'].sum()
channel_views
```

## 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

To calculate the `like_ratio`, we need to divide the number of likes by the number of dislikes ($likes / dislikes$). However, a preliminary analysis of the dataset revealed 7,387 rows where the number of dislikes is exactly 0. 



```python
# Controlliamo se ci sono video con 0 dislikes o valori mancanti
(df_clean['dislikes'] == 0).sum()
```

```python
df_clean[['likes', 'dislikes']].isna().sum()
```

We use `numpy.where()`:
- For all rows where `dislikes > 0`, the standard ratio is computed.
- For rows where `dislikes == 0`, we explicitly assign a `NaN` (Not a Number) value.

```python
import numpy as np
```

```python
df_clean['like_ratio'] = np.where(
    df_clean['dislikes'] > 0,
    df_clean['likes'] / df_clean['dislikes'],
    np.nan
)
```

## 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df_clean['publish_time'] = pd.to_datetime(df_clean['publish_time'])
```

```python
def cluster_10min_interval(timestamp):
    start_time = timestamp.dt.floor('10min') #timestamp rounded to the closet 10 minute boundary
    end_time = start_time + pd.Timedelta(minutes = 10)
    return start_time.dt.strftime('%H:%M') + ' - ' + end_time.dt.strftime('%H:%M') 
df_clean['interval'] = cluster_10min_interval(df_clean['publish_time'])
```

```python
df_clean[['publish_time','interval']]
```

## 7. For each interval, determine the number of videos, average number of likes and of dislikes.

Since a single video accumulates rows every day it remains in the trending section, performing an evaluation on the raw dataset would heavily distort the true number of unique videos published in each time slot. To prevent this bias, we drop chronological duplicates using `subset=['video_id', 'country']`, keeping only the final snapshot of each video.

```python
df_unique_analysis = df_clean.sort_values(by=['country', 'trending_date']).drop_duplicates(subset=['video_id','country'], keep='last')
```

Using the `.agg()` function, we simultaneously compute the total count of unique videos (`count`) and the mathematical averages (`mean`) for both likes and dislikes within each specific time slot.

```python
interval_analysis = df_unique_analysis.groupby('interval', as_index=False).agg(
    number_videos = ('video_id', 'count'),
    average_likes = ('likes', 'mean'),
    average_dislikes = ('dislikes', 'mean')
)
interval_analysis
```

## 8. For each tag, determine the number of videos. Notice that tags contains a string with several tags.

The `tags` column contains a structured string where multiple keywords for a single video are concatenated using the pipe delimiter (`|`). 

```python
df_unique_tags = df_clean.sort_values(by=['country', 'trending_date']).drop_duplicates(subset=['video_id','country'], keep='last')
```

We sanitize the strings by removing double quotes and applying `.str.split('|')`. This converts the flat text into a Python list of discrete tag strings.

```python
# remove unwanted characters like quotes, and split the string into a list of tags using the pipe '|' separator
df_unique_tags['tags'] = df_unique_tags['tags'].str.replace('"','').str.split('|')
```

We apply the `.explode('tags')` method. This operation transforms each element of the tag lists into a separate row, maintaining the correct relationship with the original video identifiers.

```python
# Explode the list of tags so each individual tag gets its own row
df_exploded_tags = df_unique_tags.explode('tags')
df_exploded_tags = df_exploded_tags[(df_exploded_tags['tags'] != '[none]') & ((df_exploded_tags["tags"] != ""))]
```

```python
tag_counts = df_exploded_tags.groupby('tags', as_index=False)['video_id'].count()
tag_counts.columns = ['tag', 'number_of_videos'] # rename columns
tag_counts
```

## 9. Find the tags with the largest number of videos

## 14. Read all json files with the video categories

```python
import json
import os
```

To aggregate all country categories into a single DataFrame named `totjson`, we implement an automated loop pipeline:
1. We loop through a defined list of target country codes to dynamically construct the file paths.
2. Since JSON is nested, we apply `pd.json_normalize()` directly to the `items` series. This flattens properties like `snippet.title` and `snippet.assignable` into proper tabular columns.
3. Before merging, we append a `country` column to each individual sub-table to track the geographic origin of those categories.
4. Finally, we utilize `pd.concat()` to bind all 10 DataFrames together.

```python
json_folder = "json"
countries = ['CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU', 'US']
json_df = []

for country in countries:
    file_path = os.path.join(json_folder, f"{country}_category_id.json")
    raw_json = pd.read_json(file_path)
    df_normalized = pd.json_normalize(raw_json['items'])
    df_normalized['country'] = country
    json_df.append(df_normalized)

totjson = pd.concat(json_df, ignore_index=True)
totjson
```
